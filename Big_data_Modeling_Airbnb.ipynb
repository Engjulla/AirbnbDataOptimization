```python
# I understood that we should return the city with the highes and the city with the lowest average values per property
# type.

# I used $toInt, beacuse we I wasnt able to compute the average with the original values, and with it was able to perform the calculation.
query7_1 = {
    '$group': {
        '_id': {'Property':'$property_type', 'City':'$address.market'},
        'Average_duration': {'$avg': {'$toInt': '$maximum_nights'}}
    }
}

# Group by property type and create an array named city that stores both city name and the average duration for that city.
# With the push command we create that array and assign values to the elements. With this we are able to compute the average
# for all cities per property type without having a complex primary key. Here the primary key is only the property type, and
# we can calculate the average duration of every city that has that type of property.
query7_2 = {
    '$group': {
        '_id': '$_id.Property',
        'Cities': {
            '$push': {
                'City': '$_id.City',
                'Average_duration': '$Average_duration'
            }
        },
        # Then we check the city with the maximum average, keeping that value and the name of the city, per property type.
        'City_max': {'$max': {
            'City': '$_id.City',
            'Average_duration': '$Average_duration'
            }
        },
        # We also check the city with the minimum average, keeping also that value and the name of the city, per property
        # type.
        'City_min': {'$min': {
            'City': '$_id.City',
            'Average_duration': '$Average_duration'
            }
        }
    }
}

# Finally create a query projection that will only display the property type and the city with higher average duration and
# the city with the lowest average duration for each property type
query7_3 = {
    '$project': {
        'Property Type': '$_id',
        'City with highest average duration': '$City_max.City',
        'Max Average': '$City_max.Average_duration',
        'City with lowest average duration': '$City_min.City',
        'Min Average': '$City_min.Average_duration'
    }
}


pipeline_7 = [query7_1, query7_2, query7_3]
iterator_7 = db.listingsAndReviews_new.aggregate(pipeline_7)
result_7 = list(iterator_7)

pprint('Max and Min average duration of stay: ' + str(len(result_7)))
pprint(result_7)
```

```python
# Function to calculate the average value of transactions for a property in a given period
def get_average_transaction_value(property_id, start_date, end_date):
    pipeline = [
        {'$match': {'_id': property_id}},
        {'$unwind': {'path': '$transactions', 'includeArrayIndex': 'index'}},
        {'$lookup': {
            'from': 'listingsAndReviews_new',
            'let': {'property_id': '$_id', 'index': '$index'},
            'pipeline': [
                {'$match': {'$expr': {'$eq': ['$_id', '$$property_id']}}},
                {'$unwind': {'path': '$transaction_dates', 'includeArrayIndex': 'index'}},
                {'$match': {'$expr': {'$eq': ['$index', '$$index']}}},
                {'$project': {'_id': 0, 'date': '$transaction_dates'}}
            ],
            'as': 'dates'
        }},
        {'$match': {
            'dates.date': {
                '$gte': datetime.strptime(start_date, "%Y-%m-%d"),
                '$lte': datetime.strptime(end_date, "%Y-%m-%d")
            }
        }},
        {'$group': {
            '_id': '$_id',
            'total_value': {'$sum': '$transactions'},
            'count': {'$sum': 1}
        }},
        {'$project': {
            '_id': 0,
            'average_value': {'$divide': ['$total_value', '$count']}
        }}
    ]

    result = list(db.listingsAndReviews_new.aggregate(pipeline))
    return result[0]['average_value'] if result else None
```
